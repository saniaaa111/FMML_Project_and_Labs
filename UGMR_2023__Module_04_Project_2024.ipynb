## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

# Define the function and its derivative
def f(x):
    return x**4 - 3*x**3 + 2

def df(x):
    return 4*x**3 - 9*x**2

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient

    step_size = abs(new_x - previous_x)

    x_list.append(new_x)
    slope_list.append(df(new_x))

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

# Example 3 - Divergence and Overflow

## $$h(x) = x^5 - 2x^4 + 2$$

```python
## TASK-4 : Code the same things as they were done for the first two examples

import numpy as np
import matplotlib.pyplot as plt

# Define the functions and their derivatives
def g(x):
    return x**4 - 4*x**2 + 5

def dg(x):
    return 4*x**3 - 8*x

def f(x):
    return x**2 + x + 1

def df(x):
    return 2*x + 1

# Gradient Descent
def gradient_descent(initial_x, learning_rate, precision, max_iter, func, deriv_func):
    x = initial_x
    x_list = [x]
    y_list = [func(x)]

    for _ in range(max_iter):
        gradient = deriv_func(x)
        x = x - learning_rate * gradient

        x_list.append(x)
        y_list.append(func(x))

        if abs(learning_rate * gradient) < precision:
            break

    return x_list, y_list

# Plot function and derivative side by side
x_values = np.linspace(-3, 3, 1000)

# Plot for g(x)
x_optimal, y_optimal = gradient_descent(initial_x=2, learning_rate=0.2, precision=0.00001, max_iter=50, func=g, deriv_func=dg)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_values, g(x_values), label='g(x)')
plt.scatter(x_optimal, y_optimal, color='red', label='Optimization path')
plt.title('Function g(x)')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values, dg(x_values), label="g'(x)")
plt.title("Derivative g'(x)")
plt.xlabel('x')
plt.ylabel("g'(x)")
plt.legend()

plt.tight_layout()
plt.show()

# Plot for f(x)
x_optimal, y_optimal = gradient_descent(initial_x=-2, learning_rate=0.2, precision=0.00001, max_iter=50, func=f, deriv_func=df)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_values, f(x_values), label='f(x)')
plt.scatter(x_optimal, y_optimal, color='red', label='Optimization path')
plt.title('Function f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_values, df(x_values), label="f'(x)")
plt.title("Derivative f'(x)")
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the dataset
x_5 = np.array([1, 2, 3, 4, 5])
y_5 = np.array([5, 7, 9, 11, 13])

# Define the initial thetas
thetas_initial = np.array([0.0, 0.0])

# Define the learning rate and maximum iterations
learning_rate = 0.01
max_iterations = 1000

# Define the MSE function
def mse(y, y_hat):
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc

# Gradient Descent Optimization
def gradient_descent(x, y, thetas, multiplier, max_iter):
    plot_vals = thetas.reshape(1, 2)
    mse_vals = mse(y, thetas[0] + thetas[1]*x)

    for i in range(max_iter):
        gradients = grad(x, y, thetas).mean(axis=0)
        thetas = thetas - multiplier * gradients

        # Append the new values to our numpy arrays
        plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
        mse_vals = np.append(arr=mse_vals, values=mse(y, thetas[0] + thetas[1]*x))

        # Print the progress
        if (i + 1) % 100 == 0:
            print(f'Iteration {i+1}, MSE: {mse_vals[-1]}')

    return thetas, plot_vals, mse_vals

# Define the gradient function
def grad(x, y, thetas):
    n = len(y)
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)
    return np.array([theta0_slope, theta1_slope])

# Run gradient descent
final_thetas, history_thetas, history_mse = gradient_descent(x_5, y_5.flatten(), thetas_initial, learning_rate, max_iterations)

# Results
print('Min occurs at Theta 0:', final_thetas[0])
print('Min occurs at Theta 1:', final_thetas[1])
print('MSE is:', mse(y_5, final_thetas[0] + final_thetas[1]*x_5))

# Plot the optimization process
plt.figure(figsize=(12, 4))

# Plot the data
plt.subplot(1, 2, 1)
plt.scatter(x_5, y_5)
plt.title('Data Points')

# Plot the optimization process
plt.subplot(1, 2, 2)
plt.plot(history_mse)
plt.title('MSE over Iterations')

plt.show()



```
